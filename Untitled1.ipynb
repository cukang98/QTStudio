```python
import pandas as pd

# read file URLs from files.txt
url_file = 'https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/files.txt'
file_urls = pd.read_csv(url_file, header=None)[0].tolist()

# read all CSV files and concatenate into a single dataframe
dfs = []
for file_url in file_urls:
    df = pd.read_csv(file_url)
    dfs.append(df)
grades = pd.concat(dfs)

# sort by ID and reset index
grades = grades.sort_values(by='ID').reset_index(drop=True)
```

```python
grades.head()
```

```python
# read projects.csv into a dataframe
projects = pd.read_csv('https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/projects.csv')

# read grades.csv into a dataframe (assuming you have already done this step)
# and merge with projects
grades = pd.merge(grades, projects[['ID', 'Project']], on='ID', how='left')
```

```python
grades.head()
```

```python
placements = pd.read_csv('https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/placements.csv')

# read grades.csv into a dataframe (assuming you have already done this step)
# and merge with placements
grades = pd.merge(grades, placements[['ID', 'Placement']], on='ID', how='left')
```

```python
grades.head()
```

```python
levels = pd.read_csv('https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/levels.csv')

# set Group and ID columns as a MultiIndex
grades = grades.set_index(['Group', 'ID'])
```

```python
levels = pd.read_csv('https://raw.githubusercontent.com/uqglmn/dhv/main/q3/v1/levels.csv')

# create a dictionary from levels dataframe
levels_dict = levels.set_index('Level 1').to_dict()['Level 0']
```

```python
# create a list of tuples for column MultiIndex
cols = [(levels_dict[col], col) for col in grades.columns]

# create column MultiIndex using pd.MultiIndex.from_tuples
grades.columns = pd.MultiIndex.from_tuples(cols)
```

```python
grades
```

```python
grades = grades.groupby(['Group','ID'], 
                  as_index=True).agg(lambda x:x)
grades_tomean = grades.groupby(level=[0,1]).sum()
```

```python
grades.head()
```

```python
mean_grades = grades_tomean.mean(axis=1, level=0).round(1)
```

```python
grades_withoutplacement = grades_tomean.drop('Placement', axis = 1, level = 1)
```

```python
grades_withoutplacement.head()
```

```python
mean_grades = grades_withoutplacement.mean(axis=1, level=0).round(1)
```

```python
mean_grades["Final Grade"] = grades_withoutplacement.mean(axis=1)
```

```python
mean_grades['Name'],mean_grades['Surname'] = grades['Student']['Name'], grades['Student']['Surname']
```

```python
mean_grades
```

```python
# compute the Degree Class using pd.cut
mean_grades['Class'] = pd.cut(mean_grades['Final Grade'], bins=[0, 39.5, 49.5, 59.5, 69.5, 100], labels=['Fail', 'Third', 'Lower Second', 'Upper Second', 'First'])

# sort by ID and reset index
mean_grades = mean_grades.sort_values(by='ID').reset_index()
```

```python
mean_grades
```

```python
# set Group and ID columns as a MultiIndex
mean_grades = mean_grades.set_index(['Group', 'ID'])
# create a dictionary from levels dataframe
levels_dict = {
    'Name':'Student',
    'Surname':'Student',
    'Level 4': 'Annual Grades',
    'Level 5': 'Annual Grades',
    'Level 6': 'Annual Grades',
    'Final Year': 'Annual Grades',
    'Final Grade': 'Degree',
    'Class': 'Degree',
}

# create a list of tuples for column MultiIndex
cols = [(levels_dict[col], col) for col in mean_grades.columns]

# create column MultiIndex using pd.MultiIndex.from_tuples
mean_grades.columns = pd.MultiIndex.from_tuples(cols)
```

```python
mean_grades.head()
```

```python
mean_grades = mean_grades.groupby(['Group','ID'], 
                  as_index=True).agg(lambda x:x)
```

```python
mean_grades
```

```python
# count the number of each Degree Class in mean_grades
degrees = mean_grades['Degree']['Class'].value_counts()

# convert Series to DataFrame and set index name
degrees = pd.DataFrame(degrees).rename_axis('Degree').reset_index()

# set column Degree as index
degrees = degrees.set_index('Degree')

#Save as csv 
degrees.to_csv('degrees.csv')
```
